```python
import time
import serial  # pyserial 라이브러리 임포트
import serial.tools.list_ports
```

```python
def list_serial_devices():
    """
    Lists all available serial ports on the system and prints their details.
    """
    ports = serial.tools.list_ports.comports()
    if not ports:
        print("No serial devices found.")
    else:
        print("Detected serial ports:")
        for port in sorted(ports):
            # p.device gives the port name, p.description gives a human-readable description,
            # and p.hwid gives the hardware ID (e.g., VID:PID).
            print(f"- {port.device}: {port.description} [{port.hwid}]")
            
    return ports
```

```python
list_serial_devices()
```

```python
# 시리얼 포트 설정
ser = serial.Serial(
    # port='COM3',      # Windows: COM3, Linux/Mac: '/dev/ttyUSB0'
    port='/dev/cu.usbserial-A5069RR4',      # Windows: COM3, Linux/Mac: '/dev/ttyUSB0'
    baudrate=115200,    # Baud Rate 설정 (통신 속도)
    timeout=1         # Timeout 설정 (1초 동안 데이터 없으면 종료)
)

print("시리얼 포트가 열렸습니다:", ser.is_open)  # 포트가 정상적으로 열렸는지 확인
```

```python
packets = [b'FR01801000200031804180', b'FR01801180218031804180', b'FR00001000200030004000']

for i in range(3):
    for packet in packets:
        ser.write(packet)
        time.sleep(1)  # 1초 대기
```

```python
def finger_test():
    ser.write(b'FR00001000200030004000')  # 데이터 전송
    time.sleep(0.5)  # 잠시 대기

    for i in range(0, 5, 1):
        for j in range(0, 181, 45):
            # command = f'FA{i:01d}{j:03d}02000300040005000'
            command = 'FR'
            for k in range(0, 5, 1):
                command += f'{k:01d}'
                if k == i:
                    command += f'{j:03d}'
                else:
                    command += '000'
            ser.write(command.encode())  # 데이터 전송
            # print(command)
            time.sleep(0.5)  # 잠시 대기
        # print(i)

    ser.write(b'FR00001000200030004000')  # 데이터 전송
```

```python
for _ in range(3):
    finger_test()
```

```python
ser.close()
print("시리얼 포트가 닫혔습니다:", not ser.is_open)
```
